```python
print('Reading ratings.csv file (index=false)')
ratings_columns = ['user_id', 'movie_id', 'rating', 'created_at']
ratings = pd.read_csv('D:/კურსები/FreeUni/Seminars/datasets/movies/ratings.csv', names=ratings_columns, index_col=False)
ratings
```

```python
print('Reading tags.csv file')
tags_columns = ['user_id', 'movie_id', 'tag', 'created_at']
tags = pd.read_csv('D:/კურსები/FreeUni/Seminars/datasets/movies/tags.csv', names=tags_columns, index_col=False)
tags
```

```python
print('Reading users.csv file (index specified)')
user_columns = ['user_id', 'first_name', 'last_name', 'birth_date', 'country']
users = pd.read_csv('D:/კურსები/FreeUni/Seminars/datasets/movies/users.csv', names=user_columns, index_col=False)
users
```

```python
print(''' Merge ratings with movies and users ''')
ratings_with_titles = pd.merge(movies, ratings,
                       how='left', left_on='movie_id', right_on='movie_id')

merge_result = pd.merge(ratings_with_titles,users, how='inner', left_on='user_id', right_on='user_id')
merge_result
```

```python
print(''' Join ratings with movies and users ''')
ratings_with_titles = movies.set_index('movie_id').join(ratings.set_index('movie_id'), how='inner')

join_result = ratings_with_titles.reset_index('movie_id')\
.set_index('user_id').join(users, lsuffix='_l', rsuffix='_r')

join_result
# ratings_with_titles
# join_result.reset_index()
```

```python
print('Top 20 movies which have avg rating > 4 and have been rated at least 100 times:')
top_rated = movies.set_index('movie_id')\
    .join(ratings.set_index('movie_id'), lsuffix='_mv', rsuffix='_rtg', how='inner')\

# top_rated

top_rated = top_rated[['title', 'rating']]\
        .groupby(['title'])\
        .agg(freq=('rating', 'size'), avg=('rating', 'mean'))\
        

top_rated = top_rated\
        .loc[(top_rated['freq'] >= 100) & (top_rated['avg'] >= 4)]\
        .sort_values(by=['avg', 'freq'], ascending=False)\
        .nlargest(20, 'avg')\
        .reset_index()

top_rated
```

```python
# first time only
# !pip install matplotlib

print(''' Plotting number of movies produced annualy: ''')

movies_transformed = movies.copy(deep=True)


movies_transformed['year'] = movies_transformed['title'].str[-6:].str.replace('(','').str.replace(')','')
movies_transformed['year'] = pd.to_numeric(movies_transformed['year'], errors='coerce')

movies_transformed['title'] = movies_transformed['title'].str[0:-6]

movies_transformed
# print(movies_transformed)

movies_produced_annualy = movies_transformed\
        .groupby('year')\
        .agg(num_of_movies=('title', 'size'))\
        .sort_values(by='year')\
        .reset_index()

movies_produced_annualy.plot(x='year', y='num_of_movies', legend=True, kind='line')


# movies_transformed['year'].unique()
```

```python
print('''Transform tags into list/set''')
tags_as_list = tags.groupby(['user_id', 'movie_id'])['tag'].apply(list).reset_index()

print()
print(tags_as_list)
print(type(tags_as_list))

print('''Exploding list into rows''')

print(tags_as_list.explode('tag'))
```

```python
print('''Transform multiple columns into sets''')

tags_as_set = tags\
    .drop(columns=['created_at'])\
    .groupby(['tag'])\
    .agg(set)\
    .reset_index()

tags_as_set
```

```python
print('''Explode multiple columns''')
tags_mult_col_explode = tags_as_set\
        .explode('user_id')\
        .explode('movie_id')

print(f'''Num of elements in original tags.csv: {tags.count()}''')
print(f'''Num of elements in 'exploded' tags: {tags_mult_col_explode.count()}''') # :) 
tags_mult_col_explode
```
